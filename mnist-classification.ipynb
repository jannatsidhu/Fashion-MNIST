# Firstly Testing on a Sub sample to find the best Feature Extraction method and the best Classifier

## Importing Dataset and doing Standard Scaling

```
# importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```
# getting the train and test dataset.
dataset_train = pd.read_csv("train.csv") # i am only taking train dataset and dividing it in train and test dataset so that I can get the accuracy.

X_train = dataset_train.drop(columns = ['Id','Label'])
y_train = dataset_train['Label']

X_test = dataset_train.drop(columns = ['Id','Label'])
y_test = dataset_train['Label']

# X_train and test for the testing purposes.
X_train = X_train[0:10000]
y_train = y_train[0:10000]

X_test = X_test[10001:11001].reset_index(drop=True)
y_test = y_test[10001:11001].reset_index(drop=True)
# doing pre-processing of the data.
# applying standardscaler on the dataset
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

## Evaluating different Feature Extraction method.

From above feature extraction methdods the best feature extraction method is PCA and the reason behind it is very well explained in the report.

As PCA is the best method I am using PCA in the classification.

Below you can see the code for all the feature extration methods.

### PCA

```
# applying PCA on the dataset.
from sklearn.decomposition import PCA
pca = PCA(n_components=200)# n_components=2
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# plotting the explained variance ratio
plt.plot(pca.explained_variance_ratio_)
plt.show()
v = pca.explained_variance_ratio_

# plotting the scatter plot of the PCA
cyan = y_train == 0
red = y_train == 1
blue = y_train == 2
yellow = y_train == 3
magenta = y_train == 4
plt.scatter(X_train_pca[cyan,0], X_train_pca[cyan,1],c = "cyan",s=20,edgecolor = 'k',label = 'class 0')
plt.scatter(X_train_pca[red,0], X_train_pca[red,1],c = "red",s=20,edgecolor = 'k', label = 'class 1')
plt.scatter(X_train_pca[blue,0], X_train_pca[blue,1],c = "blue",s=20,edgecolor = 'k', label = 'class 2')
plt.scatter(X_train_pca[yellow,0], X_train_pca[yellow,1],c = "yellow",s=20,edgecolor = 'k', label = 'class 3')
plt.scatter(X_train_pca[magenta,0], X_train_pca[magenta,1],c = "magenta",s=20,edgecolor = 'k', label = 'class 4')
plt.legend()
plt.title("PCA")
plt.show()
```

# Code of Catboost's ROC curve and Classification Report

## Pre-Processing

```
# importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle

# getting the train and test dataset.
dataset_train = pd.read_csv("train.csv")  
dataset_test = pd.read_csv("testX.csv")

X_train = dataset_train.drop(columns = ['Id','Label']).values
y_train = dataset_train['Label'].values

X_test = dataset_test.drop(columns = ['Id']).values

# applying standardscaler on the dataset
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

# train-test(validation) split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_train,y_train,test_size=0.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

## PCA

```
%%time
# applying PCA on the dataset.
from sklearn.decomposition import PCA
pca = PCA(n_components=200)# n_components=2
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

## Catboost algorithm

```
%%time
from catboost import CatBoostClassifier,Pool

# creating a catboosting classifier
cat_model = CatBoostClassifier(
    iterations=10, # 2020
    loss_function='MultiClass',
    eval_metric='MultiClass',
    leaf_estimation_method='Newton',
    bootstrap_type = "Bayesian",
    leaf_estimation_iterations = 100,
    random_strength = 0.5,
    l2_leaf_reg=5,
    learning_rate = 0.001,
    bagging_temperature = 0.5,
    task_type = "GPU"
)
# training the model
cat_model.fit(X_train_pca,y_train)  
```

## Classification Report

```
# classification report
from sklearn.metrics import classification_report

y_pred = cat_model.predict(X_test_pca)

report = classification_report(y_test,y_pred,labels = [0,1,2,3,4])

print(report)
```

## ROC Curve

```
# for ROC curve 
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_curve, auc


y_pred = cat_model.predict(X_test_pca)
y_pred = label_binarize(y_pred,classes=[0,1,2,3,4])

y_test = label_binarize(y_test,classes=[0,1,2,3,4])

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(5):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure()
lw = 2
plt.plot(fpr[0], tpr[0],lw=lw, label='ROC curve class 0 (area = %0.2f)' % roc_auc[0])
plt.plot(fpr[1], tpr[1],lw=lw, label='ROC curve class 1 (area = %0.2f)' % roc_auc[1])
plt.plot(fpr[2], tpr[2],lw=lw, label='ROC curve class 2 (area = %0.2f)' % roc_auc[2])
plt.plot(fpr[3], tpr[3],lw=lw, label='ROC curve class 3 (area = %0.2f)' % roc_auc[3])
plt.plot(fpr[4], tpr[4],lw=lw, label='ROC curve class 4 (area = %0.2f)' % roc_auc[4])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic Catboost')
plt.legend(loc="lower right")
plt.show()
```
